```python
import pandas as pd
df=pd.read_pickle('data/percent_change_df.pickle')
```

```python
df['TX12ma']=df['TX'].rolling(12).mean()
```

```python
print (df['TX12ma'])
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax1=plt.subplot2grid((1,1),(0,0))
df['TX'].plot(ax=ax1)
df['TX12ma'].plot(color='k',ax=ax1)
plt.show()
```

```python
df['TX12std']=df['TX'].rolling(12).std()
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((2,1), (0,0))
ax2 = plt.subplot2grid((2,1), (1,0), sharex=ax1)

df['TX'].plot(ax=ax1)
df['TX12ma'].plot(ax=ax1)
df['TX12std'].plot(ax=ax2)

plt.show()
```
