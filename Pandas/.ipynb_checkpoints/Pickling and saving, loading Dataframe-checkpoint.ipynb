## Pickling means to save any python object i.e. variables, list, dict, dataframe, ML classifier

```python
import quandl
import pandas as pd
df=quandl.get('FMAC/HPI',authtoken='YZya5Hsq6QFx8sg2V-T5')
print (df.head())
```

```python
df=df.iloc[:,:-2]
```

```python
print (df.shape)
```

## Saving the dataframe

```python
import pickle
```

```python
pickle_out=open('data/housing.pickle','wb')
pickle.dump(df,pickle_out)
pickle_out.close()
```

## Loading the pickle

```python
pickle_in=open('data/housing.pickle','rb')
loaded_df=pickle.load(pickle_in)
print (loaded_df.head())
```

## Pickling in Pandas

```python
loaded_df.to_pickle('data/housing_pandas.pickle')
loaded_df_pandas=pd.read_pickle('data/housing_pandas.pickle')
print (loaded_df_pandas.head())
```
